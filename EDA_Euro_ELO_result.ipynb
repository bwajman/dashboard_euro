```python
import pandas as pd
```

### import dataset with ELO's ranking - ELO rank before cup and after competitions

```python
df_before = pd.read_csv('/Users/admin/Downloads/data/ELO_start.csv', sep=';')
df_before.info()
```

```python
df_after = pd.read_csv('/Users/admin/Downloads/data/ELO_end.csv', sep=';')
df_after.info()
```

```python
df_after.head()
```

```python
order = ['Euro','Team','Local','Global','Rating']
df_before = df_before[order]
df_after = df_after[order]
```

```python
df_before['Team'].loc[df_before['Team']=='Czechia']= 'Czech Republic'
df_after['Team'].loc[df_after['Team']=='Czechia']= 'Czech Republic'
df_before['Team'].loc[df_before['Team']=='West Germany']= 'Germany'
df_after['Team'].loc[df_after['Team']=='West Germany']= 'Germany'
```

```python
df = pd.merge(df_before,df_after, on = ['Euro','Team'], how = 'inner')
```

```python
df
```

```python
df['Change']=df['Rating_y']-df['Rating_x']
df['Change_Global']=df['Global_y']-df['Global_x']
df['Change_Local']=df['Local_y']-df['Local_x']
```

```python
df_best = df[['Euro','Team','Change','Change_Global','Change_Local']].loc[df['Change'] == df.groupby('Euro')['Change'].transform('max')]
df_best
```

```python
df_worst = df[['Euro','Team','Change','Change_Global','Change_Local']].loc[df['Change'] == df.groupby('Euro')['Change'].transform('min')]
df_worst
```

```python
df.info()
```

```python
df_worst.to_csv('/Users/admin/Downloads/data/ELO_worst.csv')
df_best.to_csv('/Users/admin/Downloads/data/ELO_best.csv')
```

### prepare dataset with result and combine it with ELO's ranking - prepare dataset to ML algorithm

```python
result = pd.read_csv('/Users/admin/Downloads/data/result.csv',sep=';',names=['Stage','Date','Hour','Home','sep','Away','Score'])
```

```python
result.head()
```

```python
result['Home'].loc[result['Home']=='CSSR']= 'Czechoslovakia'
result['Away'].loc[result['Away']=='CSSR']= 'Czechoslovakia'
result['Home'].loc[result['Home']=='USSR']= 'Soviet Union'
result['Away'].loc[result['Away']=='USSR']= 'Soviet Union'
result['Home'].loc[result['Home']=='CIS']= 'Commonwealth of Indep States'
result['Away'].loc[result['Away']=='CIS']= 'Commonwealth of Indep States'
```

```python
result.drop(columns=['Hour','sep'],inplace=True,axis=0)
```

```python
result['Date']=result['Date'].str.split('/').str[-1]
```

```python
result.head()
```

```python
result['Date'].loc[result['Date']=='2021']='2020'
```

```python
result['Stage'].value_counts()
```

```python
result['Stage'].loc[result['Stage']=='Group 1']='Group A'
result['Stage'].loc[result['Stage']=='Group 2']='Group B'
result['Stage'].loc[result['Stage']=='Group 3']='Group C'
result['Stage'].loc[result['Stage']=='Group 4']='Group D'
```

```python
result['Stage'].value_counts()
```

```python
mapping = {'Group A' : 0, 'Group B' : 0, 'Group C' : 0, 'Group D' : 0, 'Group E' : 0, 'Group F' : 0,
          'Round of 16' : 1, 'Quarter-finals' : 2, 'Semi-finals': 3, '3rd place' : 4, 'Final':5}
```

```python
result['Stage'] = result['Stage'].map(mapping)
```

```python
result['Stage'].value_counts()
```

```python
result['Extra Time'] = result['Score'].str.contains('aet')
result['Penatly'] = result['Score'].str.contains('pso')
```

```python
result['Extra Time'] = result['Extra Time'].astype(int)
result['Penatly'] = result['Penatly'].astype(int)
```

```python
result['Score'] = result['Score'].str.replace('aet','').str.replace('pso','')
```

```python
result.head()
```

```python
result['Score']=result['Score'].str.rstrip()
```

```python
result['Score_temp']=result['Score'].str.split(' ')
```

```python
result['Score']=result['Score'].str.split(' ').str[0]
```

```python
result['H']=result['Score'].str.split(':').str[0]
result['A']=result['Score'].str.split(':').str[1]
```

```python
result
```

```python
result.drop(columns=['Score'],inplace=True,axis=0)
```

```python
result['Score_temp']=result['Score_temp'].str[1:]
```

```python
result
```

```python
result['HT_H']=result['Score_temp'].str[0].str.split(':').str[0].str.replace('(','').str.replace(',','').str.replace(' ','')
result['HT_A']=result['Score_temp'].str[0].str.split(':').str[1].str.replace(')','').str.replace(',','').str.replace(' ','')
result['FT_H']=result['Score_temp'].str[1].str.split(':').str[0].str.replace('(','').str.replace(',','').str.replace(' ','')
result['FT_A']=result['Score_temp'].str[1].str.split(':').str[1].str.replace(')','').str.replace(',','').str.replace(' ','')
result['ET_H']=result['Score_temp'].str[2].str.split(':').str[0].str.replace('(','').str.replace(',','').str.replace(' ','')
result['ET_A']=result['Score_temp'].str[2].str.split(':').str[1].str.replace(')','').str.replace(',','').str.replace(' ','')
```

```python
result
```

```python
result.drop(columns=['Score_temp'],inplace=True,axis=0)
```

```python
result.head()
```

```python
result.drop(columns=['FT_H','FT_A','ET_H','ET_A'],inplace=True,axis=0)
```

```python
result
```

```python
result['Date'] = result['Date'].astype(int)
```

```python
result.info()
```

```python
combined
```

```python
combined.drop(columns=['Date','Home'],inplace=True)
```

```python
combined['Euro'].value_counts()
```

```python
combined_copy = pd.merge(combined,df, left_on = ['Euro','Away'], right_on=['Euro','Team'], how = 'left')
combined_copy = combined_copy.rename(columns = {"Rating_x" : "Away_Rating"})
combined_copy.drop(columns=['Team','Local_x','Global_x','Local_y','Global_y','Rating_y','Change','Change_Global','Change_Local'],inplace=True)
```

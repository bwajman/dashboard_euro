```python
import pandas as pd
```

```python
df = pd.read_csv('/Users/admin/Downloads/data/players.csv',sep=',',names=['Player','Country','Matches'])
```

```python
df.head()
```

```python
df.loc[df['Player'].str[-1]=='*','Active'] = True
```

```python
df['Player']=df['Player'].str.replace('*','')
```

```python
df['Active'].fillna(False, inplace=True)
```

```python
df
```

```python
df['Player']=df['Player'].str.rstrip()
df['Player']=df['Player'].str.lstrip()
```

```python
df['Active'].fillna(False, inplace=True)
df = df.reset_index(drop=True)
df.index += 1
```

```python
df
```

```python
df.to_csv('/Users/admin/Downloads/data/players_modified.csv')
```

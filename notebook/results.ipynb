```python
import pandas as pd
```

```python
result = pd.read_csv('/Users/admin/Downloads/data/result.csv',sep=';',names=['Stage','Date','Hour','Home','#','Away','Score'])
```

```python
result = result.reset_index(drop=True)
result.index += 1
```

```python
result['Extra Time'] = result['Score'].str.contains('aet')
result['Penatly'] = result['Score'].str.contains('pso')
```

```python
result['Euro'] = result['Date'].str.split('/').str[-1]
```

```python
result['Extra Time'] = result['Score'].str.contains('aet')
result['Penatly'] = result['Score'].str.contains('pso')
```

```python
result.drop(columns = '#', axis=1, inplace=True)
```

```python
result['Score'] = result['Score'].str.replace('aet','').str.replace('pso','')
```

```python
result.head()
```

```python
result['Stage'].value_counts()
```

```python
result.loc[result['Stage'].str.contains('Group'),'Stage']='Group'
```

```python
result
```

```python
result['Score_temp']=result['Score'].str.split(' ')
result['H']=result['Score_temp'].str[0].str.split(':').str[0]
result['A']=result['Score_temp'].str[0].str.split(':').str[1]
result['HT_H']=result['Score_temp'].str[1].str.split(':').str[0].str.replace('(','')
result['HT_A']=result['Score_temp'].str[1].str.split(':').str[1].str.replace(')','')
result['HT_A']=result['HT_A'].str.replace(',','')
result['FT_H']=result['Score_temp'].str[2].str.split(':').str[0].str.replace('(','')
result['FT_A']=result['Score_temp'].str[2].str.split(':').str[1].str.replace(')','')
result['FT_A']=result['FT_A'].str.replace(',','')
result['ET_H']=result['Score_temp'].str[3].str.split(':').str[0].str.replace('(','')
result['ET_A']=result['Score_temp'].str[3].str.split(':').str[1].str.replace(')','')
result['ET_A']=result['ET_A'].str.replace(',','')
```

```python
result
```

```python
result.drop(columns='Score_temp',axis=1,inplace=True)
```

```python
result
```

```python
result.loc[result['Penatly']==True,'PEN_H'] = result['H']
result.loc[result['Penatly']==True,'PEN_A'] = result['A']
result.loc[result['Penatly']==True,'H'] = result['ET_H']
result.loc[result['Penatly']==True,'A'] = result['ET_A']
```

```python
result
```

```python
result.drop(columns=['ET_H','ET_A'],axis=1,inplace=True)
```

```python
order=['Euro','Stage','Date','Hour','Home','Away','H','A','HT_H','HT_A','Extra Time','FT_H','FT_A','Penatly','PEN_H','PEN_A']
result=result[order]
```

```python
# Penatlies without Extra Time 
result.loc[(result['Home']=='CSSR') & (result['Away']=='Italy'),'H']='1'
result.loc[(result['Home']=='CSSR') & (result['Away']=='Italy'),'A']='1'
```
